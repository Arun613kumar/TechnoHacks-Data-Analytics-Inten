# Data clean

```python
df['Age'].mean()
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
def modify_age(c):
  Age=c[0]
  Pclass=c[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 40
    elif Pclass==2:
      return 28
    else:
      return 24
  else:
    return(Age)
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(modify_age, axis=1)
```

```python
df.isnull().sum()
```

# **Change the data type of 'Age' to int**

```python
df['Age'] = df['Age'].astype(int)
```
